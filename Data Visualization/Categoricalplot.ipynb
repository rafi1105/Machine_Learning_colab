#Categorical Data Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset("tips")
data
```

##countplot

```python
sns.countplot(x=data['sex'], hue=data['smoker'])
```

##boxplot

```python
sns.boxplot(y=data['day'], x=data['tip'],palette='rainbow')
```

##violinplot

```python
sns.violinplot(x=data['day'], y=data['total_bill'],palette='rainbow')
```

##stripplot

```python
sns.stripplot(x=data['day'], y=data['total_bill'],jitter=True,palette='rainbow')
```

```python
sns.swarmplot(x=data['day'], y=data['total_bill'],palette='rainbow')
```

```python
sns.violinplot(x=data['day'], y=data['total_bill'],palette='rainbow')
sns.swarmplot(x=data['day'], y=data['total_bill'],color='black',size=3)
```

##barplot

```python
import numpy as np
sns.barplot(x=data['day'], y=data['total_bill'],estimator=np.sum) # estimator can be (mean,median,sum,std,min,max)
```

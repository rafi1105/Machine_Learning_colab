#Seaborn working

##Import

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

##Data set
distribution of work with **Quantitative **
1.   Quantitative data like (tatal_bill, tip)
2.   Quatitative or category wish (sex, smoker,day, time, size)





```python
df=sns.load_dataset('tips')
```

```python
df['size'].unique()
```

##Displot/Hisplot

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde = True, color="r") #bins no of histogram  default bins = 20
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde = True, color="g")
```

##Joint plot

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

##pairplot

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

##rugplot

```python
sns.rugplot(df['total_bill'],color='g')
```

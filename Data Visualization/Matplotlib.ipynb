# Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Mathplotlib Basic

```python
x= np.linspace(0,5,11) # (start_num,end_num,how many num between them)
x
y= x**2
y
```

```python
plt.plot(x,y, color='red',linewidth=2)
plt.title('x and y line graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

## Subplots in Matplotlib

```python
plt.subplot(2,2,1) #num_row,num_col,select num of row and col # make matrix
plt.plot(x**2,y/4,'y')
plt.subplot(2,2,4)
plt.plot(x,y,'r') # 'color'
plt.subplot(2,2,3)
plt.plot(y,x,'b')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
```

## Object orintent way

```python
fig= plt.figure(figsize=(6,4), dpi= 100) #axis the canves #control the canves #(width,height) #show more clear resulation default =100
axis1=fig.add_axes([.1,0.1,.8,.8]) # (y-axis, x-axis, width, height) calculated in %
axis1.plot(x,y)

axis2=fig.add_axes([.1,0.4,.3,.3]) #overlap
axis2.plot(x,y,'r')
```

#Matrix Plot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

##heat map
co-relation variable using prediction (-1 to 1)
<br>
- 0 then there are no relation of cols
- +1 then Both inc. continouly both
- -1 then if a_cols inc. b_cols dec.

```python
co_relation= tips[['total_bill', 'tip', 'size']]
co_relation.corr()
```

```python
sns.heatmap(co_relation.corr())
```

##Cluster map

```python
sns.clustermap(co_relation.corr())
```

##pivot table heat map

```python
pvtflight = flights.pivot_table(index='month', columns='year', values='passengers')
pvtflight
```

```python
sns.heatmap(pvtflight)
```
